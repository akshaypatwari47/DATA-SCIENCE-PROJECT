```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv("/content/drive/My Drive/DS PROJECT/train.csv", encoding = 'Latin-1')
data
```

```
data.shape
```

```
X_train = data.tweet
```

```
y_train = data.label
```

```
data.drop(['id'],axis = 1,inplace = True)
data.shape
```

```
data
```

```
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
#creating an object for CountVectorizer
count_vector = CountVectorizer()
```

```
#generate the word counts for the words in the documents
word_count = count_vector.fit(X_train) 
word_count.get_feature_names()
```

```
#Converting to a array
array = word_count.transform(X_train).toarray()
array
```

```
#Converting the array into a dataframe
df = pd.DataFrame(array,columns = word_count.get_feature_names())
df
```

```
#TFIDF Vectorizer

#Import TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
#Initialize an object for the TFIDF Vectorizer
tfidf_vect = TfidfVectorizer()
```

```
#apply the tfidf to the data(X_train)
X_train_transformed = tfidf_vect.fit_transform(X_train)
X_train_transformed
```

```
test = pd.read_csv("/content/drive/My Drive/DS PROJECT/test.csv", encoding = 'Latin-1')
test
```

```
#TFIDF Vectorizer

#Import TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
#Initialize an object for the TFIDF Vectorizer
tfidf_vect = TfidfVectorizer()
```

```
#apply the tfidf to the data(X_train)
X_train_transformed = tfidf_vect.fit_transform(X_train)
X_train_transformed
```

```
result = pd.read_csv("/content/drive/My Drive/DS PROJECT/result.csv",encoding = 'Latin-1')
result
```

```
result.drop(['id'],axis = 1,inplace = True)
result
```

```
# Apply the Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier

#initailization of object
dtree = DecisionTreeClassifier()

# Applying the classifier to the dataset
# syntax: objectname.fit(InputValues, OutputValues)
dtree.fit(X_train_transformed, y_train)
```

```
DecisionTreeClassifier()
```

```
# Predict on training data
y_train_pred = dtree.predict(X_train_transformed)
y_train_pred
```
